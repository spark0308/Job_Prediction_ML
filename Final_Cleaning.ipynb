```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df=pd.read_csv("Final_set.csv")
df.head()
```

```python
City=df.City.replace("/",",",regex=True).replace("-",",",regex=True).replace("&",",",regex=True)
df1=pd.DataFrame()
new=[]
for i in City:
    a=i.find(",")
    if(a==-1):
        new.append(i)
    else:
        new.append(i[:a])
df1["new"]=new       
df1.new1=df1.new.str.strip()
```

```python
Counter(df1.new1).most_common()
```

```python
df.City=df1.new1
df.rename(columns={"job_title": "Job_Description", "City":"Location","Eli":"Eligibility"},inplace=True)
```

```python
l=[]
for i in df["salary"].astype(float):
    l.append(i*10)
l
df["salary"]=l
df.head()
```

```python
Counter(df["Location"]).most_common()
```
